# Handling Imbalanced Class

### Here the dataset is imbalanced in a very extreme manner. There are are only 492 positive class cases as compared to over 56K negative class cases. This is <1%. To make this balanced we have two methods. SMOTE and Downsampling combimed with upweighting. SMOTE will create artificial instances to match minority class with majority class, but since the gap is too huge and even after downsampling we can have enough data we will try with Downsampling first.

```python
from imblearn.under_sampling import NearMiss

X, y = df.drop('Class', axis=1), df['Class']
nm = NearMiss()
X_bl, y_bl = nm.fit_resample(X, y)
df_bl = X_bl.copy()
df_bl['Class'] = y_bl
```

# Building Pytorch Model

```python
import torch
device = (
    "cuda"
    if torch.cuda.is_available()
    else "mps"
    if torch.backends.mps.is_available()
    else "cpu"
)
print(f"Using {device} device")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import train_test_split


# Convert 'DataFrame' objects to NumPy arrays
X_bl_array = X_bl.values
y_bl_array = y_bl.values

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_bl_array, y_bl_array, random_state=42)

# Convert the input and target arrays to tensors
train_inputs = torch.tensor(X_train, dtype=torch.float32)  # Shape: (num_samples, input_size)
train_targets = torch.tensor(y_train, dtype=torch.float32)  # Shape: (num_samples,)

test_inputs = torch.tensor(X_test, dtype=torch.float32)  # Shape: (num_test_samples, input_size)
test_targets = torch.tensor(y_test, dtype=torch.float32)  # Shape: (num_test_samples,)


# Convert the input and target tensors into a PyTorch dataset
train_dataset = TensorDataset(train_inputs, train_targets)
test_dataset = TensorDataset(test_inputs, test_targets)


# Specify the batch size for training
batch_size = 64

# Create the train loader
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

# Create the test loader
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)


```

```python
import torch.nn as nn

class BinaryClassifier(nn.Module):
    def __init__(self, input_size):
        super(BinaryClassifier, self).__init__()
        self.fc1 = nn.Linear(input_size, 32)   # Fully connected layer 1
        self.relu = nn.ReLU()                  # Activation function (ReLU)
        self.dropout = nn.Dropout(0.2)         # Dropout function      
        self.fc2 = nn.Linear(32, 1)            # Fully connected layer 2
        self.sigmoid = nn.Sigmoid()            # Activation function (Sigmoid)

    def forward(self, x):
        out = self.fc1(x)
        out = self.relu(out)
        out = self.fc2(out)
        out = self.sigmoid(out)
        return out
```

```python
def train(model, train_loader, optimizer, criterion, epochs, tolerence=0.01, warmup_period=5):
    
    train_loss = []
    train_acc = []
    

    for epoch in range(epochs):
        running_loss = 0.0
        correct = 0
        total = 0

        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)  # Move inputs and labels to the device
            optimizer.zero_grad()
            outputs = model(inputs)
            labels = labels.view(-1, 1)  # Reshape the target tensor
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

        epoch_loss = running_loss / len(train_loader)
        epoch_acc = correct / total
        train_loss.append(epoch_loss)
        train_acc.append(epoch_acc)

        # Early callback implementation with tolerence and warmup
        if epoch>warmup_period:
            t_1 = train_loss[-1]
            t_2 = train_loss[-2]
            if (t_2-t_1)<tolerence:
                break

        print(f"Epoch [{epoch+1}/{epochs}], Loss: {epoch_loss:.4f}, Accuracy: {epoch_acc:.4f}")

    # Plotting the loss and accuracy
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(range(1, len(train_loss)+1), train_loss)
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Training Loss')

    plt.subplot(1, 2, 2)
    plt.plot(range(1, len(train_acc)+1), train_acc)
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title('Training Accuracy')

    plt.tight_layout()
    plt.show()
```

```python
from sklearn.decomposition import PCA

X_bl_array = X_bl.values
y_bl_array = y_bl.values

# Perform PCA to obtain the top two principal components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_bl_array)

# Convert the PCA-transformed data to tensors
pca_inputs = torch.tensor(X_pca, dtype=torch.float32)
pca_targets = torch.tensor(y_bl_array, dtype=torch.float32)

# Convert the PCA-transformed tensors into PyTorch datasets
pca_dataset = TensorDataset(pca_inputs, pca_targets)

# Create the PCA data loader
pca_loader = DataLoader(pca_dataset, batch_size=batch_size, shuffle=False)
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.patches as mpatches
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, precision_score, recall_score


def test(model, test_loader, criterion):
    model.eval()  # Set the model to evaluation mode
    test_loss = 0.0
    test_acc = 0.0

    predictions = []  # List to store predicted values
    targets = []  # List to store actual values

    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)  # Move inputs and labels to the device

            outputs = model(inputs)
            labels = labels.view(-1, 1)  # Reshape the target tensor
            loss = criterion(outputs, labels)

            test_loss += loss.item()
            predicted = torch.round(outputs)  # Round the output to get binary predictions
            correct = (predicted == labels).sum().item()
            test_acc += correct

            # Convert tensors to CPU and append to the lists
            predictions.extend(predicted.cpu().numpy())
            targets.extend(labels.cpu().numpy())

    num_samples = len(test_loader.dataset)
    test_loss /= len(test_loader)
    test_acc /= num_samples

    print(f"Test Loss: {test_loss:.4f}, Test Accuracy: {test_acc:.4f}")
    # print(type(predictions))
    # print(*predictions)
    # print(type(targets))
    # print(*targets)
    cm = confusion_matrix(targets, predictions)
    print("Confusion Matrix:")
    # print(cm)
    cm_display = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = [0, 1])
    cm_display.plot()

    precision = precision_score(targets, predictions, pos_label=0)
    recall = recall_score(targets, predictions)
    precision_fraud_detection = cm[1][1]/cm[1].sum()
    
    # fr_precision = 

    print(f"Precision: {precision_fraud_detection:.4f}, Recall: {recall:.4f}")

    # Scatter plot of predicted and actual values using the top two PCA components
    pca_values = test_loader.dataset.tensors[0].cpu().numpy()
    plt.figure(figsize=(8, 6))

    # Create color maps for actual and predicted values
    color_map = list('green' if i==j else 'red' for i,j in zip(predictions, targets))
    # print(color_map)

    # Scatter plot of actual values
    plt.scatter(pca_values[:, 0], pca_values[:, 1], c=color_map, alpha=0.5)

    green_patch = mpatches.Patch(color='green', label='Correct Prediction')
    red_patch = mpatches.Patch(color='red', label='Incorrect Prediction')

    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.legend(handles=[green_patch, red_patch])
    plt.title('Correct & Incorrect Predictions')
    plt.show()

```

```python
import torch.optim as optim


# Create an instance of the BinaryClassifier model
input_size = X_bl.shape[1]
model = BinaryClassifier(input_size)
model = nn.DataParallel(model)
model.to(device)

# Print the model architecture
print(model)

learning_rate = 0.001
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

criterion = nn.BCEWithLogitsLoss()

train(model, train_loader, optimizer, criterion, epochs=20, tolerence=0.01, warmup_period=10)
```

```python
test(model, test_loader, criterion)
```
